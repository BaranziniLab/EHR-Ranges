Want to see if units within a same lab are in a different scale.
Brute force approach would be to make constant conversions across the board.

However, we shouldn't apply conversions to all the data at once because then we might end up with values such as 95,000,000 or 0.000000095 (i.e. cant generalize which unit to convert to across labs, since for one lab it might make more sense to have the ranges in micrograms, while in another lab it might make more sense to have the unit be in milligrams).


Need to check within each lab, which values to convert. This way we don't end up with very large or very small values, which would make us have to do another round of conversion when comparing individual results to these ranges.

The functions I implemented to accomplish this task of converting units are:

- getMaxUnit(df, c)
  - inputs a df and a cutoff value.
  - returns whether the max unit is frequent enough, the list of units present, and the max unit
- getUnits(df)
  - inputs a df
  - calls getMaxUnit(df, c)
  - returns a dictionary that maps each OMOPid to its max unit and the list of units present if a max unit is present that meats the cutoff value.
- convertUnits(df)
  - inputs a df
  - returns a df with the conversions applied

## 1. Implement functions to get conversion map (unit_map) for each lab
    a) getMaxUnit(df, c)
    b) getUnits(df)

```python

# a) getMaxUnit(df, c)
# c is the percentage cutoff we want to use to determine whether the most frequent unit is really a significant majority of the data.
# for example, at the last if statement in this code, it checks whether the amount of times the most frequent unit appeared is greater than the total number of observations * c.
# So if c = 0.7, it would only count this as the most frequent unit if it has a frequency greater or equal to 70%.
# We later experiment with different values of c to see how many clear lab ranges we can determine.
def getMaxUnit(df, c):
    # get unit counts
    counts_obj = df['unit'].value_counts()
    counts = list(counts_obj)
    maxv = 0
    tot = 0
    index = 0
    # loop through counts and determine the frequency of the most frequent unit
    for i in range(len(counts)):
        count = counts[i]
        if count > maxv:
            maxv = count
            index = i
        tot += count
    # if it is frequent enough, return True, the list of existing units in this df, and the max unit
    if maxv >= (tot*c):
        return True, list(counts_obj.index), list(counts_obj.index)[index]
    else:
        return False, None, None
```

```python
from itertools import compress

# b) getUnits(df)
# Remember, entries for the same lab tend to have different units since they come from different manufacturers, and medical professionals are also just not consistent with it within the same manufacturer.
# the purpose of this function is to take a df, and for each lab (OMOPid) find the max unit and the list of observed units used in this lab. With this, we will be able to make the conversions to the max Unit for each lab,
# depending on the max unit for this lab, solving the issue that would arise (as described earlier) if we took a brute force approach to converting units (making the same conversion regardless of the most frequent unit).
def getUnits(df):
    # get unique units for each lab
    units_obj = df.groupby('OMOPid').apply(lambda x: list(set(x['unit'])))
    units = list(units_obj)

    # get list of labs
    omops = list(units_obj.index)

    # intialize conversion dict
    omopToUnit = {}
    # get max unit info
    for i in range(len(omops)):
        if len(units[i]) < 2:
            continue
        lab = df[df['OMOPid'] == omops[i]]
        # if a max unit exists (frequency greater than 0.55), add to conversion dict
        tf, curr_units, maxUnit = getMaxUnit(lab, 0.55)
        if tf:
            omopToUnit[omops[i]] = [maxUnit, curr_units]

    return omopToUnit

unit_map = getUnits(clear_ranges)
```

# 3. Implement convertUnits(df)

```python
from statistics import mode

# converts units in a df for a lab to the most frequent unit to ensure all ranges are of same unit
# returns a the df with the conversions applied.

def convertUnits(df):
    units = list(df['unit'])
    maxUnit = mode(units)

    # df where unit is already most common
    keep = df[df['unit'] == maxUnit]

    # df where unit can be converted to the most common
    mid = df[(df['unit'].isin(conversions)) & (df['unit'] != maxUnit)]

    multipliers = []

    for unit in list(mid['unit']):
      # check if conversion exists
        if maxUnit in conversions[unit]:
          # if so, append the value for this conversion
            multipliers.append(conversions[unit][maxUnit])
        # if there is no logical conversion (ex: 'mg/24 hr' to 'mg/dl',
        # make multiplier negative to then easily filter out of df
        else:
            multipliers.append(-1)
    # apply changes.
    mid['range_low'] *= multipliers
    mid['range_high'] *= multipliers
    mid['unit'] = maxUnit

    # drop rows where range is negative, which will be those in which the multiplier was set to -1 to mark impossible conversions
    converted = mid[mid["range_high"] >= 0]


    # return df with rows converted to the most frequent unit
    return keep.append(converted)
```

# A) Merge similar ranges

```python
def merge_similar_ranges(df, ci = float):
  '''Merges similar ranges for a given lab (df), given a confidence interval (ci), in place'''
    vis = []
    range_CI = {}
    # remember to consider that one range might be mapped to another, which is mapped to a different one. Might have to merge all in this case.
    # To overcome this obstacle, get the mappings between ranges that are similar enough to merge into the same range
    for i, row in df.iterrows():
        index = row
        low = row['range_low']
        high = row['range_high']

        curr_interval = str(low)+'+'+str(high)

        # if we saw this range already, continue. If we haven't, go through and see if it falls in the CI of any visited range
        if [low, high] in vis:
            new_range = range_CI[curr_interval]
            df.at[i,'range_low'] = new_range[0]
            df.at[i,'range_high'] = new_range[1]
            continue

        # initially, map this range to itself in case it doesn't end up in the CI of another range.
        range_CI[curr_interval] = [low, high]

        # get the lower and upper bound intervals for the low and high values of the range
        low_int = [low-low*ci, low+low*ci]
        high_int = [high-high*ci, high+high*ci]

        # iterate through unique ranges we already visited
        for interval in vis:
            # if this range falls in the CI of one of the ranges we already saw
            if (interval[0] >= low_int[0]) and (interval[0] <= low_int[1]) and (interval[1] >= high_int[0]) and (interval[1] <= high_int[1]):

                #map the current range to this one
                range_CI[curr_interval] = [interval[0],interval[1]]

                # change the value of this range to match the other
                df.at[i,'range_low'] = interval[0]
                df.at[i,'range_high'] = interval[1]
            break

        if [low,high] not in vis:
            vis.append((low,high))

    return
```

# B) Merge Demographics

```python
# Input a list of ages. Returns True if 50% of age bins are present (bin_size=10 would give 0-10, 10-20, ..., 90-100),
# indicating that we can replace the age column with 'any'
def all_ages(ages, bin_size = int):
    # loop through all age bins (0-10, 10-20, ..., 90-100)
    vals = []
    for i in range(0,100,bin_size):
        lower = i
        upper = i+bin_size
        # append True if there's an age in ages list that falls in the current bin
        vals.append(any([True for age in ages if age >= lower and age < upper]))

    if (sum(vals)/len(vals)) >= 0.50:
        return True

    return False
```

```python
# Inputs a df for a lab with ranges already merged where necessary. Returns a df with 'any'
# in the demographic columns if the range does not depend on this demographic (determined on whether all possible
# values for a demographic are present for a given range). not in place.
def merge_demographics(df):

    # initialize df to append to during iteration
    output_df = pd.DataFrame(columns=list(df.columns))

    # break into new dfs per range, search to see if value can be replaced by 'any' for each range
    ranges = df[['range_low', 'range_high']].drop_duplicates()


    # iterate through unique ranges for this lab
    for i, lab_range in ranges.iterrows():
        low = lab_range['range_low']
        high = lab_range['range_high']

        # get all rows that have this specific range for this lab
        curr_df = df[(df['range_low'] == low) & (df['range_high'] == high)]


        # now check if the df for this range contains all values for each demographic and if so, change to 'any'
        # then append all rows to the new table

        # check if all values for gender are present
        curr_gender_vals = list(set(curr_df['gender']))
        if len(genders) == len(curr_gender_vals):
            curr_df['gender'] = 'any'

        # check if all values for ethnicity are present
        curr_ethnicity_vals = list(set(curr_df['ethnicity']))
        # value could be "Unknown". check if all defined ethnicities are in the current range
        if all([x in curr_ethnicity_vals for x in ethnicities]):
            curr_df['ethnicity'] = 'any'

        # check if all values for race are present
        curr_race_vals = list(set(curr_df['race']))
        # value could be "Unknown". check if all defined races are in the current range
        if all([x in curr_race_vals for x in races]):
            curr_df['race'] = 'any'

        # get age in intervals:
        curr_df['age_range'] = [ages[age//10] for age in curr_df['age']]

        # get first digit of age range to drop duplicates on later (can't drop duplicates on age_range because it's a list)
        curr_df['age_index'] = [age_range[0] for age_range in curr_df['age_range']]

        # check if all values for age (range) are present
        if all_ages(list(set(curr_df['age'])), bin_size = 30):
            curr_df['age'] = 'any'

        # drop duplicates
        curr_df.drop_duplicates(subset = (['gender', 'ethnicity', 'race', 'age_index']), inplace = True)
        # no longer need age_index col, drop it
        curr_df.drop('age_index', inplace=True, axis = 1)


        #concatenate result to output
        output_df = output_df.append(curr_df)

    return output_df
```

```python
def generateRanges(clear_ranges):
  output_df = pd.DataFrame(columns = clear_ranges.columns)

  for omopid in set(clear_ranges['OMOPid']):
      curr_df = clear_ranges[clear_ranges['OMOPid'] == omopid]

      # check if multiple units
      if (len(set(curr_df['unit'])) < 2):
          # merge similar ranges (+/- 10%)
          merge_similar_ranges(curr_df, 0.25)
          # change demographic to 'any' where applicable
          cleaned = merge_demographics(curr_df)
          # append to output
          output_df = output_df.append(curr_df)
          continue

      # check if there is a dominant unit
      if omopid not in unit_map:
          # don't append if no dominant unit
          continue

      # convert if necessary
      converted = convertUnits(curr_df)
      # merge similar ranges (+/- 10%)
      merge_similar_ranges(converted, 0.25)
      # change demographic to 'any' where applicable
      cleaned = merge_demographics(converted)
      # append to output
      output_df = output_df.append(cleaned)
```
